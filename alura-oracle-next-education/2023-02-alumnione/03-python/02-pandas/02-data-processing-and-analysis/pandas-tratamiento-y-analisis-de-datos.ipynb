## **5. Frecuencias**

### Selecciones y frecuencias

````python
datos = pd.read_csv('./data/alquiler_residencial.csv', sep = ';')
datos.head(10)
````

````python
# Seleccionar solamente los inmuebles clasificados con tipo ' Departamento'
seleccion = datos['Tipo'] == 'Departamento'
n1 = datos[seleccion].shape[0]
n1
````

````python
# Seleccionar los inmuebles clasificados con tipos 'Casa', 'Casa en condominio' y 'Casa de villa'
seleccion = (datos['Tipo'] == 'Casa') | (datos['Tipo'] == 'Casa en condominio') | (datos['Tipo'] == 'Casa de villa')
n2 = datos[seleccion].shape[0]
n2
````

````python
# Seleccionar los inmuebles con área entre 60 y 100 metros cuadrados, incluyendo los límites
seleccion = (datos['Area'] >= 60) & (datos['Area'] <= 100)
n3 = datos[seleccion].shape[0]
n3
````

````python
# Seleccionar los inmuebles que tengan por lo menos 4 cuartos y alquiler menor que $2,000.00
seleccion = (datos['Cuartos'] >= 4) & (datos['Valor'] < 2000)
n4 = datos[seleccion].shape[0]
n4
````

````python
print("N° de inmuebles clasificados con tipo 'Departamento' => {}".format(n1))
print("N° de inmuebles clasificados con tipo 'Casa', 'Casa en condominio' y 'Casa de villa' => {}".format(n2))
print("N° de inmuebles con área entre 60 y 100 metros cuadrados, incluyendo los límites => {}".format(n3))
print("N° de inmuebles que cuentan por lo menos con 4 cuartos y alquiler menor a $2,000.00 => {}".format(n4))
````

## **9. Excluyendo outliers**

### Identificando y excluyendo outliers

````python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (14, 6))
````

````python
datos = pd.read_csv('./data/alquiler_residencial.csv', sep = ';')
datos.head(10)
````

````python
datos.boxplot(['Valor'])
````

````python
datos[datos['Valor'] >= 500000]
````

````python
valor = datos['Valor']
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
````

````python
seleccion = ((valor >= limite_inferior) & (valor <= limite_superior))
````

````python
datos_new = datos[seleccion]
datos_new.boxplot(['Valor'])
````

### Identificando y excluyendo outliers por grupo

````python
datos.boxplot(['Valor'])
````

````python
datos.head(10)
````

````python
datos.boxplot(['Valor'], by = ['Tipo'])
````

````python
grupo_tipo = datos.groupby('Tipo')['Valor']
````

````python
type(grupo_tipo)
````

````python
grupo_tipo.groups
````

````python
Q1 = grupo_tipo.quantile(.25)
Q3 = grupo_tipo.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
````

````python
limite_inferior
````

````python
limite_superior
````

````python
datos_new = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
  eh_tipo = datos['Tipo'] == tipo
  eh_dento_limite = ((datos['Valor'] >= limite_inferior[tipo]) &
                   (datos['Valor'] <= limite_superior[tipo]))
  seleccion = eh_tipo & eh_dento_limite
  datos_seleccion = datos[seleccion]
  datos_new = pd.concat([datos_new, datos_seleccion])
````

````python
datos_new.boxplot(['Valor'], by = ['Tipo'])
````

````python
datos_new.to_csv('./data/alquiler_residencial_sin_outliers.csv', sep = ';', index = False)
````
